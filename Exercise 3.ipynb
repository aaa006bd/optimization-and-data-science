### Visualization of Roosenbrock function

the following function is called the Roosenbrock function

f :$R^2$ →***R***,$f(x, y) := 100(y − x^2)^2+ (1 − x)^2$

we will viusalize it and find the minimum and maximum

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = plt.gca(projection='3d')
s=0.5
x= np.arange(-2, 2.+s, s)
y=np.arange(-2, 2.+s, s)
x,y = np.meshgrid(x,y)
# z= (1.-x)**2 + 100.*(y-x*x)**2
z= (1-x)**2 + 100*(y-x**2)**2
ax.plot_surface(x, y, z)
ax.scatter(x[1][0],y[1][0],z[1][0],c='red')
```

### Visualization of Bazaraa-Shetty function

the following function is called the Bazaraa-Shetty function

$f : R^2 → R, f(x, y) := (x − 2)^4 + (x − 2y)^2$

we will viusalize it and find the minimum and maximum

```python
fig = plt.figure()
ax = plt.gca(projection='3d')
s=0.5
x= np.arange(-2, 2.+s, s)
y=np.arange(-2, 2.+s, s)
x,y = np.meshgrid(x,y)
z= (x-2)**4 +(x-2*y)**2
ax.plot_surface(x, y, z)
ax.scatter(x[1][0],y[1][0],z[1][0],c='red')
```
